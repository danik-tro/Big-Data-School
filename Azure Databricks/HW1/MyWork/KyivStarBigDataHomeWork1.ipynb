4 What are the yealiest and the latest years represented in the database?

```
import pyspark.sql.functions as F
from pyspark.sql.types import *

def find_year(text):
  text_r = text.strip(' "')[-5:-1]
  if text_r.isdigit():
    return int(text_r)
  return None
  
udf_find_year = F.udf(find_year, IntegerType())
movies_with_year = movies.withColumn("year", udf_find_year("title"))

display(movies_with_year.select(max("year").alias("MAX_year"), min(F.col("year")).alias("MIN_year")))
```

```
display(movies_with_year.orderBy("year"))
```

5 What is the minimum, average and maximum number of genres per single movie?

```
def genres_count(col):
  if col == "(no genres listed)":
    return 1
  return col.count("|") + 1

udf_genres_count = F.udf(genres_count, IntegerType())

movies_with_count_genres = movies.withColumn("number_genres", udf_genres_count("genres"))
display(movies_with_count_genres.select(max('number_genres'), min('number_genres'), avg('number_genres')))
```

### 6 For each genre, find the number of movies of that genre - The first way

```
import numpy as np
from functools import reduce

unique_values = movies.select("genres")\
          .where("genres != '(no genres listed)'")\
          .where('genres NOT LIKE "%|%"')\
          .distinct().orderBy("genres")

x = [movies.select(["movieId", "genres"])\
     .where(f'genres LIKE "%{i[0]}%"').withColumn("Genres",
                                                  udf_genres_NEW("genres" ,
                                                                 lit(i[0]))) for i in np.array(unique_values\
                                     .select("genres")\
                                     .collect()) ]
x.append(
  movies.select(["movieId", "genres"])\
     .where(f'genres LIKE "%IMAX%"').withColumn("Genres",
                                                  udf_genres_NEW("genres" ,
                                                                 lit("IMAX")))
)
movies_with_genres = reduce(lambda pre, nex: pre.union(nex), x)
```

```
movies_with_genres.groupBy("Genres")\
      .agg(count("movieId").alias("Number_of_films"))\
      .orderBy(F.col("Number_of_films").desc()).show()
```

8 What is the longest movie name in the dataset?

```
def length_title(col):
  text_r = col.strip(' "')[-5:-1]
  if text_r.isdigit():
    col = col.strip(' "')[:-6]
    
  return len(col.strip(' "'))

udf_length_title = F.udf(length_title, IntegerType())

movies.select("title")\
      .withColumn("length_title",
                  udf_length_title("title")).orderBy(F.col("length_title").desc()).show(1)
      
```

9 Show the list of reused movie names

```
def name_title(col):
  text_r = col.strip(' "')[-5:-1]
  if text_r.isdigit():
    col = col.strip(' "')[:-6]
    
  return col.strip(' "')

udf_name_title = F.udf(name_title, StringType())

repetition = movies\
  .select("title")\
      .withColumn("Name", udf_name_title("title"))\
      .groupBy("Name")\
      .agg(count("title")\
            .alias("Count_of_filmes"))\
            .orderBy(F.col("Count_of_filmes")\
                     .desc())\
                     .where("Count_of_filmes > 1")

display(repetition)
```

11 Load dataset "ratings.csv"

```
ratings = spark.read.csv("dbfs:/FileStore/shared_uploads/s.danik.tro.c@gmail.com/ratings.csv", header=True, inferSchema=True)
ratings = ratings.withColumn('date', F.from_unixtime('timestamp').cast(DateType()))
```

12 For each user, find total number of ratings provided.

```
ratings_per_user = ratings.select(["userId", 'rating'])\
      .groupBy("userId")\
      .agg(count("rating")\
      .alias("Number_of_rating"))\
      .orderBy(F.col("Number_of_rating").desc())

display(ratings_per_user)
```

13 How many movies an average user has ranked?

```
display(ratings_per_user.select(avg("Number_of_rating").alias("Average_rating_for_user")))
```

14 For each user, find the minimum, maximum and average rating he provided

```
user_ratings = ratings.select(["userId", 'rating'])\
      .groupBy("userId")\
      .agg(min("rating").alias("Min_of_rating"),
          max("rating").alias("Max_of_rating"),
          avg("rating").alias("Average_of_rating"))
display(user_ratings)
```

15 Find top 10 movies by average rating

```
inner_join = movies.join(ratings, movies.movieId == ratings.movieId)
ratings_per_film = inner_join.select(["title", "rating", "userId"]).groupBy("title")\
        .agg(avg("rating").alias("Average_rating"), count("userId").alias("Count_Voited"))\
        .orderBy(F.col("Average_rating").desc(), F.col("Count_Voited").desc(), F.col("title").asc())
display(ratings_per_film)
```

```
movs = movies_with_genres.select(F.col("movieId").alias("movieId_"), F.col("Genres").alias("genres_"))
movs_with_genres = movies.join(movs, movies.movieId == movs.movieId_).select(["movieId_", "title", "genres_"]).orderBy("title")
movs_with_genres.show()
```

```
ratings_with_genres = ratings.select(["userId", "movieId", "rating"])\
       .join(movs_with_genres, ratings.movieId == movs_with_genres.movieId_)\
       .select(["userId", "movieId", "rating", "genres_"])

```

```
final = ratings_with_genres.groupBy('userId', "genres_").agg(count("movieId").alias("Number_of_ratings"), avg("rating").alias("average_rating"))
display(final)
```
